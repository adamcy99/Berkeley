```python
import pandas as pd
import numpy as np
import pyodbc
con = pyodbc.connect('DSN=DB2v11')


daysback = 120
andWhere = """
and last_test_date > CURRENT_DATE - %d days
--and last_test_date between '2017-01-01' and CURRENT_DATE
and tw.level in ('C5')
and SUBSTR(TRIM(UPPER(dds.calcdefs)),1,4) in ('DI51')
and TRIM(UPPER(dds.corner)) in ('M7125P125AA')
and tw.family_code not in ('5S','5M')
--and trim(lot_grade) in ('6.1','6.2','6.3','6.4','6.5')
--and lot_id like ('8IQN26007%%')
--and tw.wafer_id in ('65601529SEA5')
""" % (daysback)

pullChipAPRC = """
select distinct
tw.lot_id,
tw.wafer_id,
--tw.level,
--tw.lot_grade,
--tw.family_code,
tw.last_test_date,
week(tw.last_test_date) as week,
tc.wafer_id_xy,
tc.normalized_testx,
tc.normalized_testy,
tc.normalized_testx || '/' || tc.normalized_testy as chip,
--g.geographykey,
--g.radius_center_5,
dds.corner
from 
DMIW.PattRecChipFactR prcfr
join DMIW_SYSTEMS.DERIVEDDATASETUP dds  on prcfr.derivedsetupkey = dds.derivedsetupkey
join DMIW_SYSTEMS.TESTEDWAFER tw        on prcfr.testedwaferkey = tw.testedwaferkey
join DMIW_SYSTEMS.TESTEDCHIP tc         on prcfr.testedwaferkey = tc.testedwaferkey and prcfr.testedchipkey = tc.testedchipkey
join DMIW_SYSTEMS.GEOGRAPHY g           on prcfr.geographykey = g.geographykey
where
tw.testedwaferkey is not null
%s
""" % (andWhere)
chipAPRC = pd.read_sql(pullChipAPRC,con)

chipAPRC
```

```python
pullShapeData = """
select 
--tw.wafer_id,
tc.wafer_id_xy,
dds.corner,
SUM(
CASE 
    WHEN MOD(cmsfr.wordlinelow,128) < 64 THEN 0
    ELSE 1
END) as lt64,
SUM(
CASE 
    WHEN MOD(cmsfr.wordlinelow,128) >= 64 THEN 0
    ELSE 1
END) as ge64,
COUNT(cmsfr.wordlinelow) as scfFails


from 

DMIW.CHIPMEMSHAPEFACTR cmsfr
join DMIW_SYSTEMS.DERIVEDDATASETUP dds  on cmsfr.derivedsetupkey = dds.derivedsetupkey
join DMIW_SYSTEMS.TESTEDWAFER tw        on cmsfr.testedwaferkey = tw.testedwaferkey
join DMIW_SYSTEMS.TESTEDCHIP tc         on cmsfr.testedwaferkey = tc.testedwaferkey and cmsfr.testedchipkey = tc.testedchipkey
join DMIW_SYSTEMS.GEOGRAPHY g           on cmsfr.geographykey = g.geographykey
join DMIW_SYSTEMS.PATTRECCATEGORY prc   on cmsfr.pattreccatkey = prc.pattreccatkey
join DMIW_SYSTEMS.SEGMENT s             on cmsfr.segmentkey = s.segmentkey

where
tw.testedwaferkey is not null
--and prc.categoryname not in ('REGPERF')
and prc.categoryname in ('SCF')
%s

group by 
--tw.wafer_id,
tc.wafer_id_xy,
dds.corner

with ur
""" % (andWhere)
shapeData = pd.read_sql(pullShapeData,con)

shapeData
```

```python
df = pd.merge(chipAPRC,shapeData,on=['WAFER_ID_XY','CORNER'],how='left')
df['wl64StripeYld'] = np.where((df['GE64'] / df['SCFFAILS'] > 0.95) & (df['SCFFAILS'] > 10), 0, 1)
df.to_csv("~/Downloads/64WLSCF.csv") 
```
