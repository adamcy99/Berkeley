```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
%matplotlib inline
import pyodbc
con = pyodbc.connect('DSN=ISDW')
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn import ensemble
from sklearn import metrics

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
daysback = 300

PattRecQuery = """
select left(lot_Id,5) || '_' || wafer_Id || '_' || kerf_X || '_' || kerf_Y as Lot_Wafer_X_Y,
     lot_Id, wafer_Id,
     family_Code, corner || ' ' || categoryName as parameter, count as parmValue, radius_Center_5, quadrant
from DMIW.PattRecChipFactR  pattRecChipFactR, 
     DMIW_SYSTEMS.DerivedDataSetup derivedDataSetup, 
     DMIW_SYSTEMS.PattRecCategory pattRecCategory, 
     DMIW_SYSTEMS.Geography  geography, 
     (select testedWaferKey, lot_Id, wafer_Id, family_Code, cur_Lot_Grade as lot_Grade, last_TestTimeStamp as date, level, testProgramName, last_Eqp_Id 
     from DMIW_SYSTEMS.TestedWafer testedWafer 
     where Last_test_date >= (current date - {} days) and Tech_id = '7HPP' and Calcdefs = 'I64M1' 
     group by testedWaferKey, lot_Id, wafer_Id, family_Code, cur_Lot_Grade, last_TestTimeStamp, level, testProgramName, last_Eqp_Id) testedWafer 
where testedWafer.testedWaferKey = pattRecChipFactR.testedWaferKey and 
     derivedDataSetup.derivedSetupKey = pattRecChipFactR.derivedSetupKey and 
     pattRecCategory.pattRecCatKey = pattRecChipFactR.pattRecCatKey and 
     geography.geographyKey = pattRecChipFactR.geographyKey and 
     categoryName <> '' and exclude_Flag = 'N' 
order by lot_Id, wafer_Id, normalized_TestX, normalized_TestY, corner, categoryName, level
""".format(daysback)
```

```python
df = pd.read_sql(PattRecQuery,con)
```

```python
aprctable = pd.pivot_table(df, values='PARMVALUE', index=['LOT_WAFER_X_Y'], columns=['PARAMETER'], aggfunc=np.mean).fillna(0)
```
