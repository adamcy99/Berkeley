(4) Fit a regression model that predicts accuracy from training size. What does it predict for n=60000? What's wrong with using regression here? Can you apply a transformation that makes the predictions more reasonable?

- Remember that the sklearn fit() functions take an input matrix X and output vector Y. So each input example in X is a vector, even if it contains only a single value.

```python
# Note to grader: Make sure to run question (3) before this one to have the train_sizes and accuracies lists

### STUDENT START ###

def P4():
    """
    Creates a Linear Regression that predicts accuracy from training size.
    Uses data collected in question (3).
    """
    
    # Model 1: Create a linear regression model and fit it with our data from question 3
    regr = LinearRegression()
    regr.fit(np.array(train_sizes).reshape(len(train_sizes),1), np.array(accuracies))
    
    # Use the model to predict for n = 60,000
    print("Using model1 to predict the accuracy of n = 60000, we get {:.2f}"
          .format(float(regr.predict(60000))))
    
    # Lets plot out the data points and the data points and the regression line to see what's wrong
    predictions = regr.predict(np.array(train_sizes).reshape(len(train_sizes),1))
    plt.plot(train_sizes, accuracies, "o", train_sizes, predictions, "-")
    plt.title("Model 1: Plot Showing Our Data Points and the Regression Line")
    plt.xlabel("Training Sample Size")
    plt.ylabel("Accuracy")
    plt.legend(("Data Points", "Regression Line"))
    
    # Model 2: Lets try taking the natural log of train_sizes to make the points look more linear
    log_train_sizes = np.log(train_sizes).reshape(len(train_sizes),1)
    log_regr = LinearRegression()
    log_regr.fit(log_train_sizes, np.array(accuracies))
    print("Using model2 to predict the accuracy of n = 60000, we get {:.2f}"
          .format(float(log_regr.predict(np.log(60000)))))
    logpredictions = log_regr.predict(log_train_sizes)
    plt.figure()
    plt.plot(log_train_sizes, accuracies, "o", log_train_sizes, logpredictions, "-")
    plt.title("Model 2: Doing Log Transform on train_sizes")
    plt.xlabel("ln(train_sizes)")
    plt.ylabel("Accuracy")
    plt.legend(("Data Points", "Regression Line"))
    
    # Model 3: Lets try taking the inverse of train_sizes to make the points look more linear
    inv_train_sizes = 1./np.array(train_sizes)
    inv_train_sizes = inv_train_sizes.reshape(len(train_sizes),1)
    inv_regr = LinearRegression()
    inv_regr.fit(inv_train_sizes, np.array(accuracies))
    print("Using model3 to predict the accuracy of n = 60000, we get {:.2f}"
          .format(float(inv_regr.predict(1./60000))))
    invpredictions = inv_regr.predict(inv_train_sizes)
    plt.figure()
    plt.plot(inv_train_sizes, accuracies, "o", inv_train_sizes, invpredictions, "-")
    plt.title("Model 3: Doing Inverse Transform on train_sizes")
    plt.xlabel("1/(train_sizes)")
    plt.ylabel("Accuracy")
    plt.legend(("Data Points", "Regression Line"))
    
### STUDENT END ###
P4()
```

ANSWER: Using the linear model to predict the accuracy of n = 60000, we get 1.24 which is well over 100% accuracy which is impossible. As shown in the Model 1 plot, the relationship between training sample size and acuracy resembles a logarithmic function where accuracy grows really fast for small samples and then tapers off as it hits an asymptote at accuracy = 1. It is important to note that this is not a real logarithmic function because logarithmic functions do not have a horizontal asymptote. Either way, however, the relationship between `train_sizes` and `accuracies` is very much non-linear so a linear regression won't give us a very good prediction.

As mentioned before, the relationship between `train_sizes` and `accuracies` looks like a logarithmic function, so perhaps doing a natural log transform on the x variable would yield a better linear regression. The Model 2 plot shows the result of doing this, and as you can see, the data points look to be much more linear than Model 1. However, using Model 2 to estimate the accuracy when n = 60000, we get 1.03 which is better than 1.24 that we got before, but still slightly over 100% which is still inaccurate. I think this is because a logarithmic function does not have a horizontal asymptote so as our sample sizes grow larger, we will eventually pass 100% accuracy. 

Instead of doing a natural log transform on the x variables, we can try doing an inverse transform on the x variables. This is because 1/x functions do have horizontal asymptotes. The Model 3 plot shows the result of doing this and it does look rather linear except when `train_sizes` gets very large and `1/train_sizes` approaches 0. Using Model 3 to estimate the accuracy when n = 60000, we get 0.93 which is less than 100%. However, this accuracy is also smaller than 0.937 which is our accuracy for n = 6400. Therefore, even though Model 3 is better than Model 1, it might not be better than Model 2.

Another way to approach creating a better model would be to transform the `accuracies` using a logit function which would constrain `accuracies` between 0 and 1. However, I did not attempt this so I don't know how much better the results would be.

(6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
# Note to grader: I will be using the simplified Gaussian blur algorithm given 
# in https://courses.cs.washington.edu/courses/cse140/14wi/homework/hw3/homework3.html
# However, I will weight it so that the pixel in question has a weight of 0.5 and the
# surrounding pixels combined will have a weight of 0.5.

### STUDENT START ###

def blur_algorithm(pixels, weight = 0.5):
    """Algorithm used to carry out the weighted bluring"""
    
    p1, p2, p3, p4, p5, p6, p7, p8, p9 = pixels
    return weight*p5 + (1-weight)*(p1+p2+p3+p4+p6+p7+p8+p9)/8


def blur_image(image, weight = 0.5):
    """Input a single image to blur. Outputs the blurred version of that image."""
    
    # Reshape the image of the digit we want to blur into 2D
    image = np.reshape(image, (28,28))
    # Create an empty version of image
    blurred_image = np.zeros(image.shape)
    # Iterate through each pixel of the image
    for i in range(28):
        for j in range(28):
            # Top row
            if i == 0:
                p1, p2, p3 = [0,0,0]
                p5, p8 = [image[i, j], image[i+1, j]]
                # Top left corner
                if j == 0:
                    p4, p7 = [0,0]
                    p6, p9 = [image[i, j+1], image[i+1, j+1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
                # Top right corner
                elif j == 27:
                    p6, p9 = [0,0]
                    p4, p7 = [image[i, j-1], image[i+1, j-1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
                else:
                    p4, p7 = [image[i, j-1], image[i+1, j-1]]
                    p6, p9 = [image[i, j+1], image[i+1, j+1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
            # Bottom row
            elif i == 27:
                p7, p8, p9 = [0,0,0]
                p5, p2 = [image[i, j], image[i-1, j]]
                # Bottom left corner
                if j == 0:
                    p1, p4 = [0,0]
                    p3, p6 = [image[i-1, j+1], image[i, j+1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
                # Bottom right corner
                elif j == 27:
                    p1, p4 = [image[i-1, j-1], image[i, j-1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
                    p3, p6 = [0,0]
                else:
                    p1, p4 = [image[i-1, j-1], image[i, j-1]]
                    p3, p6 = [image[i-1, j+1], image[i, j+1]]
                    blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
            # Left Column corners already taken care of
            elif j == 0:
                p1, p4, p7 = [0,0,0]
                p2, p5, p8 = [image[i-1, j], image[i, j], image[i+1, j]]
                p3, p6, p9 = [image[i-1, j+1], image[i, j+1], image[i+1, j+1]]
                blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
            # Right Column corners already taken care of
            elif j == 27:
                p3, p6, p9 = [0,0,0]
                p2, p5, p8 = [image[i-1, j], image[i, j], image[i+1, j]]
                p1, p4, p7 = [image[i-1, j-1], image[i, j-1], image[i+1, j-1]]
                blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
            # All other pixels    
            else:
                p1, p2, p3 = [image[i-1, j-1], image[i-1, j], image[i-1, j+1]]
                p4, p5, p6 = [image[i, j-1], image[i, j], image[i, j+1]]
                p7, p8, p9 = [image[i+1, j-1], image[i+1, j], image[i+1, j+1]]
                blurred_image[i][j] = blur_algorithm([p1,p2,p3,p4,p5,p6,p7,p8,p9], weight)
    #Now return the blurred image but undo the reshape
    return np.reshape(blurred_image, (784,))

def blur_data(data, weight = 0.5):
    blurred_data = np.zeros(data.shape)
    for i in range(data.shape[0]):
        blurred_data[i] = blur_image(data[i], weight)
    return np.array(blurred_data)

def plot_5images(data):
    """If plot the first 5 images in the data set"""
    plt.figure(figsize = (15,15)) # Make the plots larger
    for i in range(5): # Counter to interate through every row of plot_data
        plt.subplot(1, 5, i+1) # Create a subplot for each of the images
        plt.axis('off') # Turn off axis for aesthetics 
        # Plot each row of plot_data
        plt.imshow(np.reshape(data[i], (28,28)), cmap=plt.cm.Greys)
    plt.show()
    
def P6():
    """
    Carry out a simplified weighted Gaussian blur on each of the proposed data sets and report the accuracy.
    """

    # Preprocess the training data but not the dev data
    blurred_train = blur_data(mini_train_data, 0.5)
    # Please uncomment the next line if you want to see what the blurred images look like.
    #plot_5images(blurred_train)
    # Create a 1-NN model
    model1 = KNeighborsClassifier(n_neighbors=1)
    # Fit the model with blurred training data
    model1.fit(blurred_train, mini_train_labels)
    # Find the accuracy of the model on dev_data
    accuracy1 = model1.score(dev_data, dev_labels)
        
    print("Accuracy for blurred training data but unblurred dev data is {}".format(accuracy1))
    
    
    # Preprocess the dev data but not the training data
    blurred_dev = blur_data(dev_data, 0.5)
    # Create a 1-NN model
    model2 = KNeighborsClassifier(n_neighbors=1)
    # Fit the model with blurred training data
    model2.fit(mini_train_data, mini_train_labels)
    # Find the accuracy of the model on dev_data
    accuracy2 = model2.score(blurred_dev, dev_labels)
        
    print("Accuracy for unblurred training data but blurred dev data is {}".format(accuracy2))
    
    # Preprocess both the training and dev data
    # Create a 1-NN model
    model3 = KNeighborsClassifier(n_neighbors=1)
    # Fit the model with blurred training data
    model3.fit(blurred_train, mini_train_labels)
    # Find the accuracy of the model on dev_data
    accuracy3 = model3.score(blurred_dev, dev_labels)
        
    print("Accuracy for blurred training data and blurred dev data is {}".format(accuracy3))
    
### STUDENT END ###

P6()
```

ANSWER: It turns out if you blur the training data and then use it on unblurred development data, you get the highest accuracy. If you blur the training data and then use it on blurred development data, you get a slightly lower accuracy. If you do not blur the training data, and use it to label a blurred development data, you get a far lower accuracy.

By blurring the training data, you are decreasing the over fitting of the model towards that training data. Therefore you can get a higher accuracy with the development data.

(7) Fit a Naive Bayes classifier and report accuracy on the dev data. Remember that Naive Bayes estimates P(feature|label). While sklearn can handle real-valued features, let's start by mapping the pixel values to either 0 or 1. You can do this as a preprocessing step, or with the binarize argument. With binary-valued features, you can use BernoulliNB. Next try mapping the pixel values to 0, 1, or 2, representing white, grey, or black. This mapping requires MultinomialNB. Does the multi-class version improve the results? Why or why not?

```python
### STUDENT START ###

def P7():
    """
    Fits a Naive Bayes classifier to binarized mini_train_data and report accuracy on the dev_data
    Fits a Naive Bayes classifier with MultinomialNB and report accuracy on the dev_data
    """
    
    
    # Binary Version
    # Create BernoulliNB object and set binarize to 0.5
    clf = BernoulliNB(binarize=0.5)
    # Fit BernoulliNB model with training data
    clf.fit(mini_train_data, mini_train_labels)
    # Print out the accuracy
    print("The accuracy of the BernoulliNB is {}".format(clf.score(dev_data, dev_labels)))
    
    # 3 Variable Version
    
    # Create an empty array to hold preprocessed training data
    processed_data = np.zeros(mini_train_data.shape)
    processed_dev = np.zeros(dev_data.shape)
    
    lower_thresh = 0.33
    upper_thresh = 0.66
    
    # Go through each value and adjust them to 0, 1, 2 based on threshold (train data)
    for i in range(mini_train_data.shape[0]):
        for j in range(mini_train_data.shape[1]):
            if mini_train_data[i][j] < lower_thresh:
                processed_data[i][j] = 0
            elif mini_train_data[i][j] > upper_thresh:
                processed_data[i][j] = 2
            else:
                processed_data[i][j] = 1
                
    # Go through each value and adjust them to 0, 1, 2 based on threshold (dev data)
    for i in range(dev_data.shape[0]):
        for j in range(dev_data.shape[1]):
            if dev_data[i][j] < lower_thresh:
                processed_dev[i][j] = 0
            elif dev_data[i][j] > upper_thresh:
                processed_dev[i][j] = 2
            else:
                processed_dev[i][j] = 1
    
    # Create MultinomialNB object            
    clf2 = MultinomialNB()
    # Fit MultinomialNB model with processed training data
    clf2.fit(processed_data, mini_train_labels)
    # Print out accuracy
    print("The accuracy of the MultinomialNB is with processed dev_data is {}"
          .format(clf2.score(processed_dev, dev_labels)))
    print("The accuracy of the MultinomialNB is with unprocessed dev_data is {}"
          .format(clf2.score(dev_data, dev_labels)))
    
    plt.imshow(np.reshape(processed_data[0], (28,28)),cmap=plt.cm.Greys)
    
### STUDENT END ###

P7()
```

(10) Because Naive Bayes is a generative model, we can use the trained model to generate digits. Train a BernoulliNB model and then generate a 10x20 grid with 20 examples of each digit. Because you're using a Bernoulli model, each pixel output will be either 0 or 1. How do the generated digits compare to the training digits?

- You can use np.random.rand() to generate random numbers from a uniform distribution
- The estimated probability of each pixel is stored in feature\_log\_prob\_. You'll need to use np.exp() to convert a log probability back to a probability.

```python
### STUDENT START ###

def P10(num_examples):
    """
    Train a BernoulliNB model and then use it to plot out 20 of each digit
    """

    model = BernoulliNB(binarize=0.5)
    model.fit(mini_train_data, mini_train_labels)
    
    
    plt.figure(figsize = (15,15)) # Make the plots larger
    for digit in range(10):
        for i in range(num_examples):
            plt.subplot(num_examples, 10, i*10+digit+1) # Create a subplot for each of the images
            plt.axis('off') # Turn off axis for aesthetics 
            
            # Convert Probabilities from log form
            probabilities = np.exp(model.feature_log_prob_[digit])
            # Randomly generate pixel map for digit
            pixels = np.where((probabilities > np.random.rand(784)) == True, 1, 0)
            # Plot the digit
            plt.imshow(np.reshape(pixels, (28, 28)), cmap=plt.cm.Greys)
    
### STUDENT END ###

P10(20)
```

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
### STUDENT START ###

def P11(buckets, correct, total):
    
    """
    Considers the maximum posterior probability for each image and bucket
    them by probability to see accuracy for each bucket.
    """
    
    # Train a BernoulliNB model with a reasonable alpha value
    # I used alpha = 0.01 because earlier we found it is the most accurate
    model = BernoulliNB(alpha = 0.01, binarize= 0.5)
    model.fit(mini_train_data, mini_train_labels)
    
    # Generate the 1000x10 array for each of the dev_data images and 
    # their corresponding probabilities of being each digit
    probs = model.predict_proba(dev_data)
    
    # Go through each one of the 1000 images
    for i in range(probs.shape[0]):
        max_prob = np.max(probs[i]) 
        predicted_digit = np.argmax(probs[i])
        
        # Now figure out which bucket max_prob belongs to
        bucket_index = 0
        for j in range(len(buckets)):
            if max_prob > buckets[j]:
                bucket_index += 1
            else:
                break
        # Add to the "total" of the bucket
        total[bucket_index] += 1
        # Figure out of the prediction is correct
        if predicted_digit == dev_labels[i]:
            # If correct, add to the "correct" of the bucket
            correct[bucket_index] +=1
        
                
### STUDENT END ###

buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0): accuracy = correct[i] / total[i]
    print('p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy))
```

ANSWER: When the probability bucket increases, the accuracy of the model also increases. This suggests that this Naive Bayes model is pretty well calibrated.

```python
# Note to grader: I will try summing the pixel values in each row and each column as added features to my 

### STUDENT START ###

def add_features(data):
    """
    process the data set by summing the pixel values in each row and each column 
    and then adding them as features for each image.
    """
    # Create an array to hold processed version of the data
    # Shape should be 1000 x (784 + 28 rows + 28 cols = 840)
    proc_data = np.zeros((data.shape[0], data.shape[1] + 28 + 28))
    
    # Iterate through the data set to look at 1 image at a time
    for i in range(data.shape[0]):
        # Create array to hold processed version of the image
        # Shape should be 784 + 28 rows + 28 cols = 840
        processed = np.zeros((data.shape[1] + 28 + 28,))
        # Add the image data into the first 784 slots of processed
        processed[0:784] = data[i]
        
        # Reshape the image into 28x28 grids
        image = np.reshape(data[i], [28,28])
        
        # Go through each row of the image and calculate the sum
        row_sums = []
        for row in image:
            row_sums.append(np.sum(row))
        # Add the sums of the rows to Processed[784:812]
        processed[784:812] = np.array(row_sums)
        
        # Go through each column of the image and calculate the sum
        col_sums = []
        for col in range(image.shape[1]):
            col_sums.append(np.sum(image[:,col]))
        # Add the sum of the columns to Processed[812:840]
        processed[812:840] = np.array(col_sums)
        
        # Now add Processed image into Proc_data
        proc_data[i] = processed
        
    return proc_data
    
    
def P12():

    # Try summing the pixel values in each row and each column 
    # and adding it to data as features and adding it as features to 

    # Preprocess the training data
    proc_train = add_features(mini_train_data)
    # Preprocess the dev data
    proc_dev = add_features(dev_data)
    
    # Train a BernoulliNB model and test accuracy with unprocessed data
    model = BernoulliNB(alpha = 0.01, binarize= 0.3)
    model.fit(mini_train_data, mini_train_labels)
    accuracy = model.score(dev_data, dev_labels)
    print("The accuracy of the model without processing is {}".format(accuracy))

    # Train a BernoulliNB model and test accuracy
    model = BernoulliNB(alpha = 0.01, binarize= 0.3)
    model.fit(proc_train, mini_train_labels)
    accuracy = model.score(proc_dev, dev_labels)
    print("The accuracy of the model with processing is {}".format(accuracy))

### STUDENT END ###

P12()
```
